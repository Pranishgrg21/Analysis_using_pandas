**At what time of day, is my average sales high?**

daily total sales by time of day?

avg sales by time of day.

```python
daily_coffee_sales = data_excel.groupby(["date", "Time_of_Day"])["money"].sum().reset_index()
daily_coffee_sales
```

```python
daily_coffee_sales = data_excel.groupby(["date", "Time_of_Day"]).agg(total_sales=("money", "sum")).reset_index()
daily_coffee_sales.groupby("Time_of_Day").agg(avg_sales=("total_sales", "mean")).sort_values(by=["avg_sales"], ascending=False)
```

```python
daily_coffee_sales_df = data_excel.groupby(["date", "Time_of_Day"]).agg(total_sales=("money", "sum")).reset_index()
daily_coffee_sales_df.groupby("Time_of_Day").agg(avg_sales=("total_sales", "mean")).sort_values(by=["avg_sales"], ascending=False)
```
