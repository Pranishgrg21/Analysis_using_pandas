# Data Cleaning


```python
#  Checking missing values

print(data_csv.isnull().sum())
```

```python
#  Handle missing values (card column)

data_csv["card"] = data_csv["card"].fillna("No card")

data_csv["card"]
```

```python
#  duplicate
data_csv["card"].duplicated().sum()
```

```python
data_csv["card"].nunique()
```

```python
# Remove duplicates

print(data_csv.drop_duplicates(inplace=True))
```

```python
# Fix column names (lowercase + no spaces)

data_csv.columns = data_csv.columns.str.lower().str.replace(" ", "_")

data_csv.columns
```

```python
#  Convert date column

data_csv["date"] = pd.to_datetime(data_csv["date"], dayfirst=True)

data_csv["datetime"] = pd.to_datetime(data_csv["datetime"], dayfirst=True)

data_csv["date"]
data_csv["datetime"]
```

```python
data_csv['money'] = data_csv['money'].str.replace('R','').astype(float)
```

```python
data_csv
```

```python
#  Create new column (high value order)

data_csv["high_sales"] = data_csv["money"].apply(lambda x: "yes" if x > 30 else "No")

data_csv["high_sales"]
```

```python
data_csv[["coffee_name", "money", "high_sales"]]
```

```python
# Map example

data_csv["payment"] = data_csv["cash_type"].map({"card" : 1, "cash": 0})

print(data_csv["payment"])
print(data_csv["cash_type"])
```

```python
data_sorted = data_csv.sort_values(by="money", ascending=False)

data_sorted
```
